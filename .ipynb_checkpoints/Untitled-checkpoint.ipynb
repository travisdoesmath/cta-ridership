```python
import pandas as pd
```

```python
l_stops_df = pd.read_csv('CTA_list_of_L_stops.csv')

station_bools = l_stops_df[['MAP_ID','ADA','RED','BLUE','G','BRN','P','Pexp','Y','Pnk','O']].groupby('MAP_ID').any()

l_stations_df = l_stops_df[['MAP_ID','STATION_NAME','STATION_DESCRIPTIVE_NAME','Location']].merge(station_bools, how='left', left_on='MAP_ID', right_index=True).drop_duplicates()
l_stations_df[['latitude','longitude']] = l_stations_df['Location'].str.replace('\(|\)','', regex=True).str.split(',',expand=True).apply(pd.to_numeric)
l_stations_df.drop('Location', axis='columns', inplace=True)

l_stations_df = pd.concat([l_stations_df, pd.DataFrame([
    {
        'MAP_ID':40200,
        'STATION_NAME':'Randolph/Wabash',
        'STATION_DESCRIPTIVE_NAME':'Randolph/Wabash',
        'RED':False,
        'BLUE':False,
        'G':True,
        'BRN':True,
        'P':False,
        'Pexp':True,
        'Y':False,
        'Pnk':True,
        'O':True,
        'latitude':41.884431,
        'longitude':-87.626149
    },
    {
        'MAP_ID':40500,
        'STATION_NAME':'Washington/State',
        'STATION_DESCRIPTIVE_NAME':'Washington/State',
        'RED':True,
        'BLUE':False,
        'G':False,
        'BRN':False,
        'P':False,
        'Pexp':False,
        'Y':False,
        'Pnk':False,
        'O':False,
        'latitude':41.8837,
        'longitude':-87.6278
    },
    {
        'MAP_ID':40640,
        'STATION_NAME':'Madison/Wabash',
        'STATION_DESCRIPTIVE_NAME':'Madison/Wabash',
        'RED':False,
        'BLUE':False,
        'G':True,
        'BRN':True,
        'P':False,
        'Pexp':True,
        'Y':False,
        'Pnk':True,
        'O':True,        'latitude':41.882023,
        'longitude':-87.626098
    }
])], sort=False)

l_stations_df.to_csv('cta_l_stations.csv')

l_stations_df
```

```python
ridership_df = pd.read_csv('CTA_ridership_daily_totals.csv')

ridership_df['date'] = pd.to_datetime(ridership_df['date'])
ridership_df = ridership_df[ridership_df['station_id'] != 41580]

ridership_df.to_csv('cta_ridership.csv')

ridership_df
```

```python
df = pd.merge(ridership_df, l_stations_df, how='left', left_on='station_id', right_on='MAP_ID')
df
```

```python
def q1(x):
    return x.quantile(0.25)

def q3(x):
    return x.quantile(0.75)

rides_by_station = ridership_df.groupby('station_id').agg({
    'stationname':'first', 
    'rides':['min',q1,'mean','median',q3,'max','sum'],
    'date':['min','max']
})
rides_by_station.columns = rides_by_station.columns.map('_'.join)
rides_by_station.rename(columns={'stationname_first':'name', 'latitude_first':'latitude', 'longitude_first':'longitude'}, inplace=True)

rides_by_station.iloc[40:60]
```

```python
ridership_df['weekday'] = ridership_df['date'].dt.day_name()
weekday_ridership_df = ridership_df.groupby(['station_id', 'weekday']).mean().reset_index().pivot(index='station_id', columns='weekday', values='rides')
weekday_ridership_df.columns = weekday_ridership_df.columns.map(lambda x: 'avg_rides_' + x.lower())
weekday_ridership_df
```

```python
df = pd.merge(rides_by_station.reset_index(), l_stations_df, how='left', left_on='station_id', right_on='MAP_ID').merge(weekday_ridership_df, left_on='station_id', right_index=True)
df.to_json('static/data/cta_ridership.json', orient='records')
df
```
